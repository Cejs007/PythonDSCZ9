## Virtual environments

**<font color='#ed7d31'>PRACTICAL WORKSHOPS WITH SCREEN SHARING BY YOUR MENTOR</font>**

### Additional reading:

By default, `pip` will install packages globally for the whole system. In most cases, this is not the desired action. We can have **many different projects** on a computer, each with its own set of dependencies. It may happen that the dependencies of one project are incompatible with the dependencies of another project. Does this mean we have to keep them on separate machines?

`Virtualenv`, a **virtual environment**, comes to the rescue. You can create many virtual environments, each with its own set of dependencies kept in a place known only to that environment and not accessible to other environments.
In this way, we get insulation and our designs do not interfere with each other.
PyCharm has very good support for virtual environments. To install virtualenv you need to use
```bash
pip install virtualenv
```
and this is actually the only dependency you should install on your system globally.


The basic operation is, of course, to create a new virtual environment. The command virtualenv environment_name is used for this, where environment_name is the name of our choice.

The question is where to keep the created virtual environments. In one approach, they can all be kept together in one dedicated directory.

Another approach is to embed the virtual environment inside the project we are working on so that it becomes "self-sufficient". Usually, the environment is then called venv, a name recognized by Python programmers. The venv directory is then added to .gitignore so as not to track it in the repository. This approach is used by PyCharm, which we will discuss later.

Either way, the virtualenv tool itself has no opinion on the localization of virtual environments. It will create a new environment in the current directory and it depends on us what this directory will be.

After creating a new environment, the next step is to **activate it**. This means that we would like to be in this new environment, so that we can install packages in it and develop a project that depends on them.

In order to activate the environment, the script named activate, which is located in the virtual environment directory in the bin subdirectory, should be loaded into the current session.

Loading the script into the session is done with the Linux source command.
So if we are still in the same directory we were in when creating this environment, the correct command would be: source environment_name/bin/activate. (if we have several versions of Python on the computer, we can decide which one we want to use to create the environment by specifying the path in the -p parameter).

While activating a virtual environment does not sound very simple, verifying whether we managed to activate it is hassle-free - if we are in the middle of a virtual environment, our command line will have its name added from now until we leave the environment.

Also, pip freeze will return an empty list - there are no dependencies in the new virtual environment.

Another way to check if the virtual environment is properly activated is to run the which python (or which python3) command in Linux, which should now return the path to Python from within the directory of our virtual environment.

Once we are in the middle of a virtual environment, we can install packages normally using the pip tool.

A common situation is one where we created the environment because we want to start working on some project that exists on GitHub, let's say we want to get involved in some open source project.

In such a situation, we will create a clean virtual environment, activate it and clone the project from GitHub. Usually, such a project will have a requirements.txt file, so we will execute the command
```bash
pip install -r requirements.txt
```
, which in our clean virtual environment will install all dependencies necessary to run the project.

It may also happen that we are the main developers working on the project in a virtual environment, and another developer wants to join. Then we will check the dependencies in the project by running the
```bash
pip freeze > requirements.txt
```
command and we will share the resulting file with it so that it can easily recreate the environment.


Finally, there comes a point where we want to start working on a different project with a different set of dependencies.

If we want to switch from one virtual environment to another, we first have to leave the one we are in. Use the `deactivate` command to do this.

If we do it correctly, the environment name will disappear from our command line, pip freeze will show globally installed packages, and which python will point to python installed on the system or configured in the session.

### File open modes

The `.open()` function has the following modes:
* `"r"` - read only (default)
* `"w"` - for writing, if the file does not exist it will be created, if it does exist the old contents will be deleted (over-written)
* `"x"` - for creating, the operation will fail if the file already exists
* `"a"` - appending, if the file does not exist, it will be created, if it exists, new content will be appended to the end of the file (you will not loose previous content)
* `"t"` - text mode (default)
* `"b"` - binary mode
* `"+"` - means that when opened, the cursor will move to the end of the file. By default it is set at the beginning.

The modes can be used together, e.g. `a + b` means opening a binary file for writing, but the file, if it exists, will not be shortened, and appending will be done at the end of the file.

The last parameter of the `.open()` function is `encoding`, ie. file encoding (see more [text encoding](https://www.puredevsoftware.com/blog/2017/11/17/text-encoding/)). The default depends on the operating system, `cp1252` on Windows and `utf-8` on Linux.

### Opening the file

After you finish working on the file, each time, it needs to be closed. This is done with the `.close()` method called on an object of the File type.

It's worth noting that the code below is not secure:

````python
f = open('/content/drive/MyDrive/SDA Data Science/files/sentences.txt')

# do some stuff

f.close()
````

The best way to open a file is to use the file in a context using `with` keyword. The `with` keyword opens the context found in the indented block below it. When we leave the indented context block, the file will be automatically closed, so we don't have to close it ourselves.

This eliminates the confusion caused by forgetting to close the file when not using the context manager and it also protects us in a situation when errors occur in the code and the file is not closed.

````python
with open('/content/drive/MyDrive/SDA Data Science/files/sentences.txt') as f:
    pass
    # do some stuff

# file is automatically closed by now :)
````

### Reading from file

While working with **<font color='#5b9bd5'>text files</font>**, we very often want to **<font color='#5b9bd5'>read the file line by line</font>**.

There are three main ways to read a file line by line;
* the prehistoric approach
    * using the `.readline()` method
    ```python
    file_path = '/content/drive/MyDrive/SDA Data Science/files/sentences.txt'

    f = open(file_path)

    while(True):
        line = f.readline()
        if not line:
            break
            f.close()
        print(line)
    ```
* the medieval approach - using the `.readlines()` method
    * `.readlines()` returns a list which elements are consecutive single lines read from the file
    ```python
    file_path = '/content/drive/MyDrive/SDA Data Science/files/sentences.txt'

    with open(file_path) as f:
        lines = f.readlines()
        for line in lines:
            print(line)
    ```
    > It can still be useful when we don't want to read line by line, but to navigate through the file the way we chose. Note, however, that readlines will read the entire file, regardless of its size.
* the standard approach - use a lazy iterator returned by an object of `File` type.
    * the file descriptor itself returns an iterator that passes line by line
    > typically this is better than `.readlines()` because it does not read the entire file into the memory. This method is preferred.

````python
file_path = '/content/drive/MyDrive/SDA Data Science/files/sentences.txt'

with open(file_path) as f:
    for line in f:
        print(line)
        break
````

````python
def filter_out_outliers(df, col, cut_perc=2):
  upper_lim = df[col].quantile(1-cut_perc/100)
  lower_lim = df[col].quantile(cut_perc/100)

  return df[(df[col] < upper_lim) & (df[col] > lower_lim)]
````

````python
filter_out_outliers(chipotle, 'item_price [$]', 2)
````

### Scaling

In most cases, the numerical characteristics of a dataset are not specific and vary from one another. In fact, it doesn't make sense to expect the age and income columns to have the same range. But from a machine learning perspective, the same scope helps improve the model.

Scaling solves this problem. Continuous functions become range identical after the scaling process. This process is not mandatory for many algorithms, but still worthwhile. However, distance-based algorithms such as k-NN or k-Means must have scaled continuous functions as model inputs.
Basically, there are two common ways to scale: normalization and standardization.
